```python
#Em resumo:
# O código lê um arquivo CSV, adiciona 3 colunas com valores constantes:
# Horiz.Res - Resistividade Horizontal,
# Vert.Res - Resistividade Vertical e
# CA - coeficiente de anisotropia
# depois  salva o DataFrame modificado como um novo arquivo CSV.
# Adicionei ao nome do arquivo original (_cols) indicando que novas colunas estão presentes.

import pandas as pd

# Caminho do arquivo original
input_path = '/content/Jubarte_JUB20_2002H_74.1V_98.5.csv'

# Carregar o arquivo CSV no pandas DataFrame
df = pd.read_csv(input_path)

# Valores extraídos do nome do arquivo
horiz_res_value = 74.1
vert_res_value = 98.5
coeficiente_anisotropia_value = 1.33

# Inserir as novas colunas no DataFrame
df['Horiz.Res'] = horiz_res_value
df['Vert.Res'] = vert_res_value
df['CA'] = coeficiente_anisotropia_value

# Caminho para salvar o novo dataset
output_path = '/content/Jubarte_JUB20_2002H_74.1V_98.5_cols.csv'

# Salvar o DataFrame modificado em um novo arquivo CSV
df.to_csv(output_path, index=False)

print(f"Novo arquivo salvo em: {output_path}")
```

```python
# Este código lista os primeiros 10 arquivos CSV em um diretório, carrega cada um deles em um DataFrame,
# e então concatena todos esses DataFrames em um único. Por fim, ele salva o DataFrame resultante
# em um novo arquivo CSV e imprime a localização desse arquivo.

import pandas as pd
import glob

# Caminho onde os arquivos estão localizados
input_path = '/content/'

# Usar glob para listar os primeiros 10 arquivos .csv do diretório
csv_files = sorted(glob.glob(input_path + "*.csv"))[:10]

# Inicializar uma lista para armazenar os DataFrames
df_list = []

# Ler e concatenar cada arquivo CSV
for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

# Concatenar todos os DataFrames em um único DataFrame
concatenated_df = pd.concat(df_list, ignore_index=True)

# Caminho para salvar o arquivo concatenado
output_path = '/content/Jubarte_Line20_2_concat.csv'

# Salvar o DataFrame concatenado em um novo arquivo CSV
concatenated_df.to_csv(output_path, index=False)

print(f"Arquivos concatenados e salvos em: {output_path}")
```
